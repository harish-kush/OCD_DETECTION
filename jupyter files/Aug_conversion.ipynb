```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_parquet('../datasets/ocd_dataset.parquet')
display(df.head())
```

```python
df['EEG_Electrode_1'][0].dtype
```

```python
electrode_cols = [col for col in df.columns if 'EEG_Electrode_' in col]
```

```python
eeg_data_reshaped = []

for index, row in df.iterrows():
    # Extract EEG data for all electrode columns in the current row
    # Each entry in `electrode_cols` contains a list of 256 floats
    current_eeg_sample = np.array([row[col] for col in electrode_cols])

    # Transpose the array to get shape (timesteps, features)
    # From (32 electrodes, 256 timesteps) to (256 timesteps, 32 electrodes)
    current_eeg_sample_transposed = current_eeg_sample.T

    eeg_data_reshaped.append(current_eeg_sample_transposed)

# Convert the list of 2D arrays into a single 3D NumPy array
eeg_data_reshaped = np.array(eeg_data_reshaped)

print(f"Shape of the reshaped EEG data: {eeg_data_reshaped.shape}")
```

```python
from sklearn.preprocessing import StandardScaler

# Identify non-EEG feature columns
categorical_cols = ['sex', 'education']
numerical_cols = ['age', 'IQ', 'EQ']

# One-hot encode categorical features
df_encoded = pd.get_dummies(df[categorical_cols], columns=categorical_cols, drop_first=True)

# Initialize and apply StandardScaler to numerical features
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[numerical_cols]), columns=numerical_cols)

# Combine all processed non-EEG features
non_eeg_features_processed = pd.concat([df_encoded, df_scaled], axis=1)

print("Head of processed non-EEG features:")
print(non_eeg_features_processed.head())
print(f"\nShape of non_eeg_features_processed: {non_eeg_features_processed.shape}")
```

```python
non_eeg_features_processed[['sex_M', 'education_University']] = non_eeg_features_processed[['sex_M', 'education_University']].astype(int)

non_eeg_np = non_eeg_features_processed.values
non_eeg_expanded = np.repeat(non_eeg_np[:, np.newaxis, :], 256, axis=1)

X_combined = np.concatenate([eeg_data_reshaped, non_eeg_expanded], axis=2)
X_combined.shape
```

```python
y = df['disorder/diagnosis'].apply(lambda x: 1 if x == 'Obsessive compulsive disorder' else 0)
```

```python
X_ocd = X_combined[y == 1]
X_non_ocd = X_combined[y == 0]
```

```python
print("OCD shape:", X_ocd.shape)
print("Non-OCD shape:", X_non_ocd.shape)
```

```python
def augment_eeg(sample):
    aug_samples = []
    
    # Noise
    noise = np.random.normal(0, 0.01, sample.shape)
    aug_samples.append(sample + noise)
    
    # Scaling
    scale = np.random.uniform(0.9, 1.1)
    aug_samples.append(sample * scale)
    
    # Time shift
    shift = np.random.randint(1, 10)
    aug_samples.append(np.roll(sample, shift, axis=0))
    
    return aug_samples
```

```python
target_ocd = 20000
X_aug = []

while len(X_ocd) + len(X_aug) < target_ocd:
    for sample in X_ocd:
        augmented = augment_eeg(sample)
        X_aug.extend(augmented)
        
        if len(X_ocd) + len(X_aug) >= target_ocd:
            break

X_aug = np.array(X_aug[:target_ocd - len(X_ocd)])

X_ocd_final = np.concatenate([X_ocd, X_aug], axis=0)
y_ocd_final = np.ones(X_ocd_final.shape[0])

print("Final OCD:", X_ocd_final.shape)
```

```python
target_non_ocd = 30000

indices = np.random.choice(len(X_non_ocd), target_non_ocd, replace=False)
X_non_ocd_final = X_non_ocd[indices]
y_non_ocd_final = np.zeros(target_non_ocd)

print("Final Non-OCD:", X_non_ocd_final.shape)
```

```python
X_final = np.concatenate([X_ocd_final, X_non_ocd_final], axis=0)
y_final = np.concatenate([y_ocd_final, y_non_ocd_final], axis=0)
```

```python
idx = np.random.permutation(len(X_final))
X_final = X_final[idx]
y_final = y_final[idx]
```

```python
X_flat = X_final.reshape(X_final.shape[0], -1)

df_final = pd.DataFrame(X_flat)
df_final['label'] = y_final

print("Final Dataset Shape:", df_final.shape)
```

```python
columns = [
    'no.', 'sex', 'age', 'education', 'IQ', 'EQ', 'disorder/diagnosis'
] + [f'EEG_Electrode_{i}' for i in range(1, 33)]

data = []

for i in range(len(X_final)):
    row = {}
    
    # ---- Non-EEG ----
    row['no.'] = i
    
    # NOTE: y_final → label (0/1)
    row['disorder/diagnosis'] = y_final[i]
    
    # agar encoded hai toh approx reverse ya direct le lo
    non_eeg = X_final[i][0, 32:]  # static features
    
    # adjust indices based on your encoding
    row['age'] = non_eeg[0]
    row['IQ'] = non_eeg[1]
    row['EQ'] = non_eeg[2]
    
    # categorical (example)
    row['sex'] = non_eeg[3]
    row['education'] = non_eeg[4]
    
    # ---- EEG ----
    eeg = X_final[i][:, :32]  # (256, 32)
    
    for j in range(32):
        row[f'EEG_Electrode_{j+1}'] = eeg[:, j]  # full time-series
    
    data.append(row)

df_final_struct = pd.DataFrame(data, columns=columns)

print(df_final_struct.head())
```
