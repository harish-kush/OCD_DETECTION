## EDA Summary

Based on the Exploratory Data Analysis steps performed, here's a summary of the key findings:

*   **Missing Values**: We identified that 'Unnamed: 122' has 100% missing values, while 'education' had 1.59% and 'IQ' had 1.37% missing values. We addressed the missing values in 'education' and 'IQ' by imputing them with their respective medians.
*   **Duplicated Values**: There are no duplicated rows in the dataset.
*   **Feature Types**: The dataset contains 1145 numerical features and 4 categorical features, which are 'sex', 'eeg.date', 'main.disorder', and 'specific.disorder'.
*   **Numerical Feature Distributions**: We visualized the distributions of 'age', 'education', and 'IQ' using histograms, providing insights into their spread and any potential skewness.
*   **Categorical Feature Counts**: Count plots were generated for the categorical features, showing the frequency of each category within 'sex', 'eeg.date', 'main.disorder', and 'specific.disorder'.
*   **Correlations**: A correlation matrix was displayed for 'age', 'education', and 'IQ', illustrating the relationships between these selected numerical features.

```python
from scipy.stats import shapiro

stat, p = shapiro(df['IQ'])

print("Statistic:", stat)
print("P-value:", p)

if p > 0.05:
    print("Normal Distribution")
else:
    print("Not Normal")
```

```python
from scipy.stats import shapiro

stat, p = shapiro(df['education'])

print("Statistic:", stat)
print("P-value:", p)

if p > 0.05:
    print("Normal Distribution")
else:
    print("Not Normal")
```

```python
from scipy.stats import ttest_ind

male = df[df['sex']=='M']['IQ']
female = df[df['sex']=='F']['IQ']

stat, p = ttest_ind(male, female)

print("P-value:",p)
```

```python
from scipy.stats import f_oneway

groups = [group['IQ'].values
          for name, group in df.groupby('main.disorder')]

stat, p = f_oneway(*groups)

print("P-value:",p)
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

table = pd.crosstab(df['sex'], df['main.disorder'])

stat, p, dof, expected = chi2_contingency(table)

print("P-value:",p)
```

```python
from scipy.stats import pearsonr

corr, p = pearsonr(df['age'], df['IQ'])

print("Correlation:",corr)
print("P-value:",p)
```

```python
from scipy.stats import kruskal

groups = [group['IQ'].values
          for name, group in df.groupby('main.disorder')]

stat, p = kruskal(*groups)

print("P-value:",p)
```
